## Sesión 1. Sistemas numéricos

### Notación de construcción del conjunto

Supongamos que tenemos una colección de números $0,1,2,3,4,\dots,9$. Cada uno de ellos representa un dígito y podemos denotar el conjunto de dígitos con la letra $D$. Es decir,

$$
 D = {0,1,2,3,4,\cdots,9}.
$$

Utilizando la notación de construcción, podemos reescribir lo anterior como sigue

$$
 D = \underbrace{\{ \underbrace{x}_{todas\ las} \quad \underbrace{|}_{tal\ que} \quad x\ es\ un\ dígito \}}_{conjunto}
$$

#### Ejemplos

1. $E = \{ x|x\ es\ un\ dígito\ par\} = \{0,2,4,6,8\}$
2. $O = \{ x|x\ es\ un\ dígito\ impar\} = \{1,3,5,7,9\}$


### Números naturales

Este conjunto de números está definido como:

$$
 \mathbb{N} = \left\{ 0, 1, 2, \dots \right\}.
$$
y cumple con las siguientes propiedades: 

1. Asociativa
2. Conmutativa
3. Modulativa

con las operaciones: 

1. Adición
2. Multiplicación.

### Números enteros

Cumplen las mismas propiedades que los números naturales y está definico como:

$$
 \mathbb{Z} = \left\{\dots, -3, -2, -1, 0, 1, 2, 3, \dots \right\}.
$$

Además, cumple con la propiedad del inverso aditivo

$$
 a + (-a) = -a + a = 0.
$$


### Números racionales

Cumplen con las propiedades de los números enteros además de la propiedad del inverso multiplicativo

$$
 a\cdot \frac{1}{a} = \frac{1}{a}\cdot a = 1, \quad si~ a\neq 0.
$$

Este conjunto de números está definido como:

$$
 \mathbb{Q} = \left\{ \frac{a}{b}~ |~ a \in \mathbb{Z}~ y ~ b \in \mathbb{Z}; b \neq 0 \right\}.
$$

Los números racionales se pueden representar en forma decimal con finitas o infinitas cifras decimales. Por ejemplo:

* $\frac{1}{4} = 0.25$

* $\frac{1}{7} = 0.142857142857\dots$

* $\frac{1}{2} = 0.5$

* $\frac{17}{100} = 0.17$

Existen números que **no pueden ser representados como el cociente de dos números**. Esto se debe a que contienen infinitas cifras decimales **no periódicas**. Por ejemplo:

* $0.10110011100011110000\dots$

* $\sqrt{2} = 1.414213562373095$

Estos números no son racionales y son clasificados dentro de un nuevo conjunto llamado números irracionales $\left( \mathbb{I} \right)$.

#### Ejemplo

Obtener la longitud de la diagonal de de un cuadrado de lado 1. Esta medida equivale a encontrar $x$ tal que $x^{2}=2$.

![longitud_cuadrado]()

* $x>0.$
* Despejamos $x$
    * $x^{2}=2,$
    * $\sqrt{x^{2}} = \sqrt{2},$
    * $\left(x^{2} \right)^{1/2} = \sqrt{2},$
    * $\left(x \right)^{2/2} = \sqrt{2},$
    * $x = \sqrt{2}.$

**Nota:** Otros números relacionales son: $\pi$, $e$, $2\sqrt{3}$, $1-\sqrt{3}$.

### Números reales

Este conjunto de números se puede representar por decimales finitas o infinitas periódicas $\left( \mathbb{Q} \right)$ o infinitas no periódicas $\left( \mathbb{I} \right)$ junto con las operaciones de suma y multiplicación. Este conjunto se denota como $\mathbb{R}$.

Estos números se representan geométricamente sobre una recta llamada **recta real**.

![recta_reales]()

El conjunto de los números reales es dividido en el $0$ de la siguiente manera:

* *Reales positivos*
$$
 \mathbb{R}^{+} = \left\{ x \in \mathbb{R}~ | ~ x > 0 \right\}.
$$

* *Reales negativos*
$$
 \mathbb{R}^{-} = \left\{ x \in \mathbb{R}~ | ~ x < 0 \right\}.
$$

## Operaciones básicas

En álgebra es común utilizar letras tales como $x$, $y$, $a$, $b$ y $c$ para representar números. La siguiente tabla incluye las operaciones que se pueden realizar con estas representaciones.

| Operación      | Símbolo       | Significado   |
|:--------------:|:-------------:|:-------------:|
| Suma           | $a+b$         | $a$ más $b$   |
| Resta          | $a-b$         | $a$ menos $b$ |
| Multiplicación | $a\cdot b$    | $a$ por $b$   |
| División       | $\frac{a}{b}$ | $a$ entre $b$ |

Consideraciones:

1. Evite utilizar números mixtos. Por ejemplo, $$2\frac{3}{4}$$

### Orden de operaciones

| Operador                  | Símbolo          | Nivel de prioridad |
|:-------------------------:|:----------------:|:------------------:|
| Paréntesis                | ( )[ ]{ }        | Mayor              |
| Potencias y raíces        | $\wedge~\sqrt{}$ |                    |
| Multiplicación y división | $\cdot~ /$       |                    |
| Suma y resta              | $+~-$            | Menor              |

### Leyes de los signos

$$
 a(-b) = -(ab)
$$

$$
 (-a)b = -(ab)
$$

$$
 (-a)(-b) = ab
$$
  
$$
 -(-a) = a
$$

$$
 \frac{a}{-b} = \frac{-a}{b} = -\frac{a}{b}
$$

$$
 \frac{-a}{-b} = \frac{a}{b}
$$

### Propiedades de cancelación

$$
 ac = bc~ \Rightarrow ~a = b, \quad si~ c\neq 0.
$$

$$
 \frac{ac}{bc} = \frac{a}{b} \quad si~ b\neq 0,~ c\neq 0.
$$

## Propiedades de los números reales

* **Axioma 1. Propiedad conmutativa**

$$
 x + y  = y + x, \quad xy = yx.
$$

* **Axioma 2. Propiedad asociativa**

$$
 x + (y + z) = (x+y) + z, \quad x(yz) = (xy)z.
$$

* **Axioma 3. Propiedad distributiva**

$$
 x(y + z) = xy + xz.
$$

* **Axioma 4. Existencia de elementos neutros**

Existen dos números reales distintos, 0 y 1 tales que para cada número real $x$ se tiene

$$
 0 + x = x + 0 = x \quad y \quad 1\cdot x = x \cdot 1 = x.
$$

* **Axioma 5. Existencia de negativos**

Para cada número real $x$ existe un número real $y$ tal que 

$$
 x + y = y + x = 0.
$$

* **Axioma 6. Existencia del recíproco**

Para cada número real $x \neq 0$ existe un número real $y$ tal que 

$$
 xy = yx = 1.
$$

De los axiomas anteriores se deducen las leyes del Álgebra elemental.

Sean $a,b,c,d \in \mathbb{R}$ entonces

* **Teorema 1. Ley de simplificación para la suma**

Si $a+b = a+c$ entonces $b=c$.

* **Teorema 2. Posibilidad de la sustracción**

Dados $a$ y $b$ existe uno y sólo un $x$ tal que $a + x = b$. $x$ está designado por $b-a$. Si $0-a$ se escribe como $-a$ e indica el negativo de $a$.

* **Teorema 3.**

$$
 b - a = b + (-a).
$$

* **Teorema 4.**

$$
 -(-a) = a.
$$

* **Teorema 5.**

$$
 a(b - c) = ab - ac.
$$

* **Teorema 6.**

$$
 0\cdot a = a\cdot 0 = 0.
$$

* **Teorema 7. Ley de simplificación para la multiplicación**

Si $ab = ac$ y $a\neq 0$ entonces $b=c$.

* **Teorema 8. Posibilidad de la división**

Dados $a$ y $b$ con $a \neq 0$, existe uno y sólo un $x$ tal que $ax = b$. Donde $x$ está definida por $b/a$ y se denomina como cociente de $b$ y $a$. Además, $1/a$ puede escribirse como $a^{-1}$ y se define como recíproco de $a$.

* **Teorema 9.**

Si $a\neq 0$, entonces $b/a=b \cdot a^{-1}$.

* **Teorema 10.**

Si $a\neq 0$, entonces $(a^{-1})^{-1} = a$.

* **Teorema 11.**

Si $ab = 0$ entonces, $a=0$ o $b=0$.

* **Teorema 12.**

$$
 (-a)b = -(ab)~ y ~ (-a)(-b) = ab.
$$

* **Teorema 13.**

$$
 \frac{a}{b} + \frac{c}{d} = \frac{ad + bc}{bd} ~ si ~ b\neq 0 ~ y ~ d \neq 0.
$$

* **Teorema 14.**

$$
 \frac{a}{b}\frac{c}{d} = \frac{ac}{bd} ~ si ~ b \neq 0 ~ y ~ d \neq 0.
$$

* **Teorema 15.**

$$
 \frac{a}{b}\frac{c}{d} = \frac{ad}{bc} ~ si ~ b \neq 0, c\neq 0 ~ y ~ d\neq 0.
$$

## Sesión 2. Signos de agrupación

Los signos de agrupación se dividen en cuatro clases:

1. Paréntesis ordinario $(~)$
2. Corchetes $[~]$
3. Llaves $\{~\}$
4. Vínculo o barra $\overline{~~~~}$

Indican que las cantidades entre ellos deben ser considerados como **una sola cantidad**. 

Considere las siguientes expresiones

$$
 a + (b - c) = a + b - c.
$$

$$
 x + (-2y + z) = x - 2y + z.
$$

$$
 a - (b + c) = a - b - c.
$$

$$
 x - (-y + z) = x + y - z.
$$

### Reglas para manipular los signos de agrupación

1. Si hay un $+$ antes de los signos de agrupación, entonces se deja el signo de cada una de las cantidades dentro de éstos.
2. Si hay un $-$ antes de los signos de agrupación, entonces se cambia el signo a cada una de las cantidades dentro de éstos.

#### Ejemplo. Simplifique las siguientes expresiones

$$
 a + [b - c] + 2a - (a + b)
$$

$$
 m + (4n-6) + 3m - (n + 2m -1)
$$

## Operaciones con fracciones

Una **fracción algebraica** está definida como el cociente de dos expresiones algebraicas. Es decir $\frac{a}{b}$ es una fracción algebraica porque está representado como el cociente de la expresión $a$ (dividendo) y la expresión $b$ (divisor).

En una fracción algebraica, el dividendo $a$ se llama **numerador** y el divisor $b$ **denominador**.

Existen dos tipos de expresiones algebraicas:

1. Algebraica **entera**. Aquella expresión cuyo denominador es 1.

    $$
     a = \frac{a}{1}; \quad x + y = \frac{x+y}{1}.
    $$
    
2. Algebraica **mixta**. Consta de una parte entera y una fraccionaria.

    $$
     a + \frac{b}{c}; \quad x - \frac{3}{x-a}.
    $$

### Aritmética del cociente

$$
 \frac{a}{b} + \frac{c}{d} = \frac{ad}{bd} + \frac{bc}{bd} = \frac{ad + bc}{bd}, \quad si~ b\neq 0,~d\neq 0.
$$

$$
 \frac{a}{b}\cdot \frac{c}{d} = \frac{ac}{bd}, \quad si~ b\neq 0,~d\neq 0.
$$

$$
 \frac{\frac{a}{b}}{\frac{c}{d}} = \frac{a}{b}\cdot \frac{d}{c} = \frac{ad}{bc}, \quad si~b\neq 0,~c\neq 0,~d\neq 0.
$$


### Suma, resta, multiplicación y división de cocientes

a. $$\frac{2}{3} + \frac{5}{2} = \frac{2\cdot 2}{3\cdot 2} + \frac{3 \cdot 5}{3\cdot 2} = \frac{2\cdot 2 + 3\cdot 5}{3\cdot 2} = \frac{4+15}{6} = \frac{19}{6}.$$

b. $$\frac{3}{5} - \frac{2}{3} = \frac{3}{5} + \left( -\frac{2}{3} \right) = \frac{3}{5} + \frac{-2}{3}$$
    
    Aplique la regla de los signos y la aritmética del cociente para obtener -1/15.

c. $$\frac{8}{3}\cdot \frac{15}{4}$$
    
    Aplique las propiedas de cancelación y la aritmética del cociente para obtener 10.

d. $$\frac{\frac{3}{5}}{\frac{7}{9}}$$

### Mínimo común múltiplo (MCM) para sumar dos fracciones

Encuentre $\frac{8}{15}$ + $\frac{5}{12}$. 

Para ello, primero debemos encontrar los múltiplos de los denominadores.

$$
 15,30,45,\textbf{60},75,90,105,120,\dots
$$

$$
 12,24,36,48,\textbf{60},72,84,96,108,120,\dots
$$

Aquí el MCM para $15$ y $12$ es $\textbf{60}$. Entonces

$$
 \frac{8}{15} + \frac{5}{12} = \frac{8}{15}\cdot \frac{4}{4} + \frac{5}{12}\cdot \frac{5}{5} = \frac{32}{60} + \frac{25}{60} = \frac{32+25}{60} = \frac{57}{60}.
$$

## Sesión 3. Exponentes y sus propiedades

Los exponentes enteros permiten representar la multiplicación repetida de un número real, es decir:

$$
 3^{4} = 3 \cdot 3 \cdot 3 \cdot 3 = 81.
$$

Sea $a$ un número real y $n$ un entero positivo, entonces $a^{n}$ representa el producto de $n$ factores de $a$

$$
 a^{n} = \underbrace{a\cdot a\cdot \cdots \cdot a}_{factores\ n},
$$
donde $a=a^{1}$.

En la expresión anterior, $a$ es conocido como la **base** y $n$ el **exponente** o **potencia**.

$$
 4\cdot 3^{2} = 4\cdot 9 = 36.
$$

$$
 2^{2} + 3^{2} = 4 + 9 = 13.
$$

$$
 -2^{4} = -16.
$$

$$
 5\cdot 3^{2} + 2\cdot 4 = 5 \cdot 9 + 2\cdot 4 = 45 + 8 = 53.
$$

Ten en cuenta que los paréntesis pueden marcar la diferencia

$$
 (-2)^{4} = (-2)(-2)(-2)(-2) = 16.
$$

$$
 (2+3)^{2} = 5^{2} = 25
$$

### Propiedades y leyes de exponentes

$$
 a^{0} = 1, \quad si~ a\neq 0.
$$

$$
 a^{-n} = \frac{1}{a^{n}}, \quad si~ a\neq 0.
$$

$$
 a^{m}a^{n} = a^{m+n}.
$$

$$
 \left(a^{m} \right)^{n} = a^{mn}
$$

$$
 \left(ab\right)^{n} = a^{n}b^{n}
$$

$$
 \frac{a^{m}}{a^{n}} = a^{m-n} = \frac{1}{a^{n-m}},\quad si~a \neq 0.
$$

$$
 \left( \frac{a}{b} \right)^{n} = \frac{a^{n}}{b^{n}},\quad si~b\neq 0.
$$

### Raíces cuadradas

La raíz cuadrada es la operación inversa de elevar al cuadrado un número real. Esta operación está denotada por el símbolo $\sqrt{~}$. Por ejemplo, $\sqrt{25} = 5$.

#### Notas importantes
1. Los números negativos no tienen raíces cuadradas (números reales).
2. La raíz cuadrada de $0$ es $0$.
3. La raíz cuadrada de un número positivo es positiva.
4. Si $c\geq 0$, entonces $\left( \sqrt{c} \right)^{2} = c$.

$$
 \sqrt{\frac{1}{16}} = \frac{1}{4}
$$

$$
 \left( \sqrt{1.4} \right)^{2} = 1.4
$$

#### Propiedades

$$
 \sqrt{a^{2}} = |a|
$$

$$
 \sqrt[n]{a} = a^{\frac{1}{n}}
$$

$$
 \sqrt[n]{a^{m}} = a^{\frac{m}{n}}
$$

### Simplificación de expresiones

$$
 2^{-3} = \frac{1}{2^{3}} = \frac{1}{8}
$$

$$
 x^{-3}\cdot x^{5} = x^{-3+5} = x^{2}, \quad x \neq 0.
$$

$$
 \left( x^{-3} \right)^{2} = x^{-3\cdot 2} = x^{-6} = \frac{1}{x^{6}}, \quad x \neq 0.
$$

$$
 \left( \frac{2}{3} \right)^{4} = \frac{2^{4}}{3^{4}} = \frac{16}{81}.
$$

$$
 \frac{x^{-2}}{x^{-5}} = x^{-2}\cdot x^{-5} = x^{-2-(-5)} = x^{3}, \quad x \neq 0.
$$

$$
 \left( \frac{x^{-3}}{3y^{-1}} \right)^{-2} = \frac{\left(x^{-3} \right)^{-2}}{\left(3y^{-1} \right)^{-2}} = \frac{x^{6}}{3^{-2}\left( y^{-1} \right)^{-2}} = \frac{x^{6}}{\frac{1}{9}y^{2}} = \frac{9x^{6}}{y^{2}}.
$$

## Sesión 4. Problemas de abstracción
1. La suma de las edades de A y B es 84 años, y B tiene 8 años menos que A. Hallar ambas edades.

    a. Sea $x=$ la edad de A.
    
    b. B tiene 8 años menos que A. Entonces,
        
    $$
     x-8=edad\ de\ B.
    $$
    
    c. La suma de las edades de A y B es 84 años
    
    $$
     x + (x-8) = 84
    $$
    
    d. Simplificando, tenemos
    
    $$
     x + x = 84 + 8,
    $$
    
    $$
     2x = 92,
    $$
    
    $$
     x = \frac{92}{2} = 46 ~(edad\ de\ A)
    $$
    
    La edad de B está dada por 
    
    $$
     x-8=46-8=38.
    $$
    
    
2. Pagué $\$87$ por un libro, un traje y un sombrero. El sombrero costó $\$5$ más que el libro y $\$20$ menos que el traje. ¿Cuánto pagué por cada cosa?

    a. Sea $x=$ el precio del libro.
    
    b. El sombrero costó $\$5$ más que el libro
    
    $$
     x + 5 = precio\ del\ sombrero
    $$
    
    c. El sombrero costó $\$20$ menos que el traje; el traje costó $\$20$ más que el sombrero
    
    $$
     (x+5) + 20 = x + 25 ~(precio\ del\ sombrero)
    $$
    
    d. Pagué $\$87$ por un libro, un traje y un sombrero.
    
    $$
     x + (x + 5) + (x + 25) = 87
    $$
    
    e. Simplificando, tenemos
    
    $$
     3x + 30 = 87,
    $$
    
    $$
     3x = 87 - 30,
    $$
    
    $$
     3x = 57,
    $$
    
    $$
     x = \frac{57}{3} = 19~(precio\ del\ libro),
    $$
    
    $$
     x + 5 = 19 + 5 = 24~(precio\ del\ sombrero),
    $$
    
    $$
     x + 25 = 19 + 25 = 44~(precio\ del\ traje).
    $$
    
    
3. La suma de tres números enteros consecutivos es 156. Hallar los números.

    a. Sea $x=$ número menor
    
    b. $x+1$ el segundo número consecutivo
    
    c. $x+1$ el tercer número consecutivo
    
    d. Entonces
    
        $$
         x + (x+1) + (x+2) = 156,
        $$
        
        $$
         3x + 3 = 156,
        $$
        
        $$
         3x = 156 - 3,
        $$
        
        $$
         x = \frac{153}{3} = 51~(primer\ número),
        $$
        
        $$
         x + 1 = 51 + 1 = 52~(segundo\ número),
        $$
        
        $$
         x + 2 = 51 + 2 = 53~(tercer\ número).
        $$
