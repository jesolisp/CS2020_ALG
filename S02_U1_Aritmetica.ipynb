## Sesión 2. Signos de agrupación

Los signos de agrupación se dividen en cuatro clases:

1. Paréntesis ordinario $(~)$
2. Corchetes $[~]$
3. Llaves $\{~\}$
4. Vínculo o barra $\overline{~~~~}$

Indican que las cantidades entre ellos deben ser considerados como **una sola cantidad**. 

Considere las siguientes expresiones

$$
 a + (b - c) = a + b - c.
$$

$$
 x + (-2y + z) = x - 2y + z.
$$

$$
 a - (b + c) = a - b - c.
$$

$$
 x - (-y + z) = x + y - z.
$$

### Reglas para manipular los signos de agrupación

1. Si hay un $+$ antes de los signos de agrupación, entonces se deja el signo de cada una de las cantidades dentro de éstos.
2. Si hay un $-$ antes de los signos de agrupación, entonces se cambia el signo a cada una de las cantidades dentro de éstos.

#### Ejemplo. Simplifique las siguientes expresiones

$$
 a + [b - c] + 2a - (a + b)
$$

$$
 m + (4n-6) + 3m - (n + 2m -1)
$$

## Operaciones con fracciones

Una **fracción algebraica** está definida como el cociente de dos expresiones algebraicas. Es decir $\frac{a}{b}$ es una fracción algebraica porque está representado como el cociente de la expresión $a$ (dividendo) y la expresión $b$ (divisor).

En una fracción algebraica, el dividendo $a$ se llama **numerador** y el divisor $b$ **denominador**.

Existen dos tipos de expresiones algebraicas:

1. Algebraica **entera**. Aquella expresión cuyo denominador es 1.

    $$
     a = \frac{a}{1}; \quad x + y = \frac{x+y}{1}.
    $$
    
2. Algebraica **mixta**. Consta de una parte entera y una fraccionaria.

    $$
     a + \frac{b}{c}; \quad x - \frac{3}{x-a}.
    $$

### Aritmética del cociente

$$
 \frac{a}{b} + \frac{c}{d} = \frac{ad}{bd} + \frac{bc}{bd} = \frac{ad + bc}{bd}, \quad si~ b\neq 0,~d\neq 0.
$$

$$
 \frac{a}{b}\cdot \frac{c}{d} = \frac{ac}{bd}, \quad si~ b\neq 0,~d\neq 0.
$$

$$
 \frac{\frac{a}{b}}{\frac{c}{d}} = \frac{a}{b}\cdot \frac{d}{c} = \frac{ad}{bc}, \quad si~b\neq 0,~c\neq 0,~d\neq 0.
$$


### Suma, resta, multiplicación y división de cocientes

a. $$\frac{2}{3} + \frac{5}{2} = \frac{2\cdot 2}{3\cdot 2} + \frac{3 \cdot 5}{3\cdot 2} = \frac{2\cdot 2 + 3\cdot 5}{3\cdot 2} = \frac{4+15}{6} = \frac{19}{6}.$$

b. $$\frac{3}{5} - \frac{2}{3} = \frac{3}{5} + \left( -\frac{2}{3} \right) = \frac{3}{5} + \frac{-2}{3}$$
    
    Aplique la regla de los signos y la aritmética del cociente para obtener -1/15.

c. $$\frac{8}{3}\cdot \frac{15}{4}$$
    
    Aplique las propiedas de cancelación y la aritmética del cociente para obtener 10.

d. $$\frac{\frac{3}{5}}{\frac{7}{9}}$$

### Mínimo común múltiplo (MCM) para sumar dos fracciones

Encuentre $\frac{8}{15}$ + $\frac{5}{12}$. 

Para ello, primero debemos encontrar los múltiplos de los denominadores.

$$
 15,30,45,\textbf{60},75,90,105,120,\dots
$$

$$
 12,24,36,48,\textbf{60},72,84,96,108,120,\dots
$$

Aquí el MCM para $15$ y $12$ es $\textbf{60}$. Entonces

$$
 \frac{8}{15} + \frac{5}{12} = \frac{8}{15}\cdot \frac{4}{4} + \frac{5}{12}\cdot \frac{5}{5} = \frac{32}{60} + \frac{25}{60} = \frac{32+25}{60} = \frac{57}{60}.
$$
