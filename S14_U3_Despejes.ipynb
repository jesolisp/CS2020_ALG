### Ejemplo 1

Un triángulo rectángulo tienen un área de $40 cm^{2}$ y su hipotenusa es $2cm$ más larga que uno de los catetos.

1. Si $x$ denota la longitud de ese cateto, demuestre que $2x^{3} + x^{2} - 3600 = 0$.
2. Pruebe que la ecuación anterior sólo tiene una raíz positiva.
3. Encuentre las dimensiones del triángulo rectángulo y compruebe los resultados obtenidos.

### Soluciones

1. Si $x$ denota la longitud de ese cateto, demuestre que $2x^{3} + x^{2} - 3600 = 0$.

#### Solución propuesta por Juan José

$$
 (x-12)\left(2x^2+23x+300\right)=0
$$

Para encontrar las raíces tenemos

$$
 x-12 = 0 \rightarrow x = 12
$$

Del segundo tenemos

$$ 
 2x^2 + 23x+300 = 0
$$

donde 

$$
 a = 2, \quad b = 23, \quad c = 300, \quad
$$

#### Nota:
En el método de la cuadrática, al radicando se le conoce como **discriminante** y éste se representa por

$$
 \triangle = b^{2} - 4ac.
$$

$$
 \triangle = (23)^{2} - 4(2)(300) \rightarrow \triangle = 529 - 2400 \rightarrow \triangle = -1871
$$


**Consideraciones**

1. Si $\triangle > 0$, entonces hay dos soluciones reales diferentes.
2. Si $\triangle = 0$, entonces hay una solución real de multiplicidad dos.
3. Si $\triangle < 0$, entonces no hay soluciones reales.

**¿Cómo calculamos el área de un triángulo rectángulo?**

$$
 Area = \frac{b\cdot h}{2}
$$

Como conozco $Area$ sustituyo

$$
 40 = \frac{b\cdot h}{2}
$$

Recordando al buen Pitágoras

$$
 C^{2} = A^{2} + B^{2} \rightarrow C = \sqrt{A^{2} + B^{2}}
$$

Considerando que la hipotenusa es $2cm$ más larga que uno de los catetos, entonces

$$
 x + 2 \rightarrow 12 + 2 = 14 \rightarrow C = 14
$$

y también conozco a uno de los catetos, entonces

$$
 A = 12
$$

como conozco $A$ y $C$, despejo $B$ del teorema del buen Pitágoras

$$
 B^{2} = C^2 - A^2 = \rightarrow B = \sqrt{C^2 - A^2} \rightarrow B = \sqrt{196-144} \rightarrow B = \sqrt{52} \rightarrow B = 2\sqrt{13}
$$

### Ejemplo 2

La base de un rectángulo es el doble de la altura disminuida en $5$. Además, su perímetro es mayor o igual que $72$. Encuentre las dimensiones del rectángulo.

La base está dada como

$$
 b = 2a-5
$$

y el perimétro

$$
 2b + 2a = 72
$$

Sustituyendo, tengo

$$
 2(2a-5) + 2a = 72 \rightarrow 4a - 10 + 2a = 72
$$

Reduciendo términos semejantes, tenemos

$$
 6a = 62
$$

Despejamos $a$

$$
 a = \frac{62}{6} = \frac{31}{3}
$$

Y sustituimos en 

$$
 b = 2\cdot\frac{31}{3}-\frac{15}{3} \rightarrow b = \cdot\frac{62}{3} - \frac{15}{3} \rightarrow b = \frac{47}{3}
$$

## Ejemplo 3
La suma de dos números es 15 y su producto es 26. ¿Cuáles son dichos números? 

* Dos números
* El resultado de sumarlos es 15
* El resultado de multiplicarlos es 26

$$
 x + y = 15 \rightarrow x = 15 - y
$$

$$
 xy = 26
$$

Sustituimos para tener

$$
 (15 - y)y = 26
$$

Eliminamos los signos de agrupación

$$
 15y - y^2 = 26
$$

Multiplicamos por $-1$ y acomodamos la igualdad

$$
 y^{2} - 15y + 26 = 0
$$

Aplicando el método del tanteo, necesitamos

* dos números que sumados me den $-15$
* y multiplicaods me den $26$

$$
 (y-13)(y-2) = 0
$$

Encontrando las raíces de la ecuación, tenemos

$$
 y - 13 = 0 \rightarrow y = 13
$$

y

$$
 y - 2 = 0 \rightarrow y = 2
$$

## Ejemplo4
Halla dos números cuya suma es 14 y la de sus cuadrados 100.

* Dos números
* Sumados me debe dar 14
* La suma de sus cuadrados me debe dar 100

Sea $a$ el primer número, y $b$ el segundo, entonces

$$
 a + b = 14,
$$

$$
 a^2 + b^2 = 100,
$$

Despejamos $a$ de la primera igualdad

$$
 a = 14 - b,
$$

Sustituimos el resultado en la segunda

$$
 (14 - b)^2 + b^2 = 100
$$

Desarrollamos y obtenemos

$$
 14^2 - 2(14)(b) + b^2 + b^2 = 100 \rightarrow 196 - 28b + b^2 + b^2 = 100
$$

reducimos términos semejantes

$$
 2b^2 - 28b + 96 = 0
$$

Multiplicamos lo anterior por $\frac{1}{2}$

$$
 \frac{1}{2} \left(2b^2 - 28b + 96  \right) = \frac{1}{2} \left(0 \right) \rightarrow b^2 - 14b + 48 = 0
$$

Aplicando el método del tanteo

$$
 (b-8)(b-6) = 0
$$

Encontramos las raíces

$$
 b - 8 = 0 \rightarrow b = 8
$$

y 

$$
 b - 6 = 0 \rightarrow b = 6
$$
