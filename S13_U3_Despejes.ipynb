## Estructuras algebráicas complejas

Una **expresión racional algebraica** está definida como el cociente de dos polinomios tal que

$$
 a(x) = \frac{p(x)}{q(x)}, con\ q(x) \neq 0.
$$

Por ejemplo

$$
 a(x) = \frac{3x^{2} + 2}{3^{2} - 6}.
$$

También existen expresiones cuyos términos son primos entre sí. A éstas se les conoce como **expresión algebráica reducida**. Por ejemplo

$$
 \frac{x^{2} + 7x + 12}{x^{2} - 9} = \frac{(x+3)(x+4)}{(x+3)(x-3)} = \frac{x+4}{x-3}
$$

## Operaciones con expresiones racionales algebráicas

Las expresiones racionales algebráicas se pueden sumar y restar de la siguiente manera

### Con el mismo denominador

$$
 \frac{x^{3}}{x^{3} + 1} + \frac{3x^{2} + 1}{x^{3} + 1} + \frac{3x}{x^{3} + 1} = \frac{x^{3} + 3x^{2} + 3x + 1}{x^{3} + 1} = \frac{(x+1)^{3}}{x^{3} + 1}.
$$


### Con distinto denominador

$$
 \frac{1}{x + 1} + \frac{x + 2}{x^{2} - x + 1} = \frac{x^{2} - x + 1 + x^{2} + 3x + 2}{x^{3} + 1} = \frac{2x^{2} + 2x + 3}{x^{3} + 1}.
$$

También se pueden multiplicar. Para ello se aplica el mismo procedimiento que para los números racionales.

$$
 \frac{3x + 2}{x^{2}-4}\cdot \frac{x + 2}{3x^{2} + 5x + 2} = \frac{3x+2}{(x+2)(x-2)}\cdot \frac{x+2}{(3x + 2)(x+1)} = \frac{1}{(x-2)(x+1)}
$$

### Ejemplo

Sutano y Perengano pueden pintar su carro en 4 días. Sutano, por su experiencia, puede hacer el trabajo en 6 días menos del tiempo en que lo puede hacer Perengano. ¿Cuánto tiempo le toma a cada uno hacerlo?


#### Planteamiento

Sea $x$ el tiempo de Perengano y $x-6$ el de Sutano. Entonces, en 1 día realizan

$$
 \frac{1}{x}, \quad \frac{1}{x-6},
$$
respectivamente.

Juntos, en un día pintan

$$
 \frac{1}{x} + \frac{1}{x-6}
$$

Dado que ambos pueden pintar su carro en 4 días, entonces tenemos

$$
 4\cdot \left[\frac{1}{x} + \frac{1}{x-6}\right] = 1
$$

#### Solución

$$
 \frac{1}{x} + \frac{1}{x-6} = \frac{1}{4}
$$

Manipulando

$$
 \frac{2x-6}{x^{2}-6x} = \frac{1}{4}
$$

Multiplicamos cruzado

$$
 4\cdot [2x - 6] = x^{2}-6x
$$

Multiplicamos

$$
 8x-24 = x^{2} - 6x
$$

Reduciendo términos semejantes, tenemos

$$
 x^{2} - 6x - 8x + 24 = 0
$$

Y finalmente

$$
 x^{2} + 14x - 24 = 0
$$

Si recordamos la solución por factorización

$$
 (x - 12)(x - 2) = 0
$$

Tenemos como soluciones

$$
 x - 12 = 0 \rightarrow x = 12,
$$

y

$$
 x - 2 = 0 \rightarrow x = 2.
$$

## Expresión irracional algebráica

Se le conoce a todas aquellas expresiones algebráicas que contienen radicales. Su representación es

$$
 a\cdot \sqrt[n]{p(x)}
$$
donde $a$ denota al coeficiente, $\sqrt[n]{p(x)}$ el radical con $n \in \mathbb{N}$ y $p(x)$ definido como el radicando.

### Propiedades y leyes de exponentes

$$
 \begin{aligned}
  a^{0} &= 1, \quad si\ a\neq 0. \\
  a^{-n} &= \frac{1}{a^{n}}, \quad si\ a\neq 0. \\
  a^{m}a^{n} &= a^{m+n}. \\
  \left(a^{m} \right)^{n} &= a^{mn} \\
  \left(ab\right)^{n} &= a^{n}b^{n} \\
  \frac{a^{m}}{a^{n}} = a^{m-n} &= \frac{1}{a^{n-m}},\quad si\ a \neq 0. \\
  \left( \frac{a}{b} \right)^{n} &= \frac{a^{n}}{b^{n}},\quad si\ b\neq 0. \\
 \end{aligned}
$$

### Propiedades de las raíces

$$
 \begin{aligned}
  \sqrt{a^{2}} &= |a|. \\
  \sqrt[n]{a} &= a^{\frac{1}{n}}. \\
  \sqrt[n]{a^{m}} &= a^{\frac{m}{n}}.
 \end{aligned}
$$

### Ejemplo

Simplifique

$$
 2\cdot \sqrt[4]{x^{5}y^{6}z^{7}} = 2\cdot \sqrt[4]{x\cdot x^{4}\cdot y^{2}\cdot y^{4}\cdot z^{3}\cdot z^{4}} = 2\cdot \sqrt[4]{xy^{2}z^{3}} \cdot \sqrt[4]{x^{4}y^{4}z^{4}} = 2xyz\cdot \sqrt[4]{xy^{2}z^{3}} 
$$

## Operaciones con radicales

1. **Términos semejantes**. Son aquellas expresiones irracionales que contienen la parte radical como **semejante**.

### Ejemplo

Considere la siguiente ecuación y haga la reducción por términos semejantes

$$
 2\cdot \sqrt[3]{x^{4}} + 3\cdot \sqrt[3]{x^{7}y^{6}} = 2\cdot \sqrt[3]{x\cdot x^{3}} + 3\cdot \sqrt[3]{x^{3}\cdot x^{4}\cdot y^{3}\cdot y^{3}} = 2\cdot \sqrt[3]{x}\sqrt[3]{x^{3}} + 3\cdot \sqrt[3]{x}\cdot \sqrt[3]{x^{3}\cdot x^{3}\cdot y^{3} \cdot y^{3}} 
$$

Simplificando tenemos

$$
 2\cdot \sqrt[3]{x}\sqrt[3]{x^{3}} + 3\cdot \sqrt[3]{x}\cdot \sqrt[3]{x^{3}\cdot x^{3}\cdot y^{3} \cdot y^{3}} = \sqrt[3]{x}\left(2x + 3x^{2}y^{2} \right)
$$

2. **Multiplicación.** Se realiza cuando las expresiones tienen el mismo índice de raíz. Por ejemplo

$$
 4\sqrt[3]{x^{2}} \cdot 3 \sqrt[3]{a\cdot y} = 12 \sqrt[3]{a\cdot x^{2} y}
$$

Otro ejemplo puede ser

$$
 3\sqrt[3]{x^{2}} \cdot 2 \sqrt[2]{x^{5}} = 3 x^{\frac{2}{3}} \cdot 2 x^{\frac{5}{2}} = 6\cdot x^{\frac{19}{6}} = 6 \sqrt[6]{x^{19}}
$$

Recordemos el **MCM**

$$
 \begin{aligned}
  3, \textbf{6}, 9, 12, 15 \\
  2, 4, \textbf{6}, 8, 10
 \end{aligned}
$$

Entonces

$$
 \frac{2}{3} + \frac{5}{2} = \frac{2}{3}\cdot \frac{2}{2} + \frac{5}{2}\cdot \frac{3}{3} = \frac{4 + 15}{6} = \frac{19}{6}
$$

3. **División.**

$$
 \frac{4x}{\sqrt{3x} + \sqrt{2x^{3}}} = \frac{4x}{\sqrt{3x} + \sqrt{2x^{3}}} \cdot \frac{\sqrt{3x} - \sqrt{2x^{3}}}{\sqrt{3x} - \sqrt{2x^{3}}} = \frac{4x\left( \sqrt{3x} - \sqrt{2x^{3}} \right)}{3x - 2x^{3}}
$$

## Ejercicios de clase

Resuelva las siguientes ecuaciones y verifique su solución

1. Despeje $g$ de $P=\pi\sqrt{\frac{e}{g}}$
2. Despeje $x$ de $y=\frac{1}{\sqrt{1-x}}$

Simplifique las siguientes expresiones

* $\sqrt[3]{x\sqrt{x}}$
* $\frac{\sqrt{x} - \frac{1}{\sqrt{x}}}{\frac{1-\sqrt{x}}{\sqrt{x}}}$

1. Despeje $g$ de $P=\pi\sqrt{\frac{e}{g}}$

Lo puedo expresar como

$$
 P = \pi \left( \frac{e}{g} \right)^{\frac{1}{2}}
$$

Elevamos al cuadrado ambos lados de la igualdad y tenemos

$$
 P^{2} = \left( \pi \left( \frac{e}{g} \right)^{\frac{1}{2}}\right)^{2}
$$

$$
 P^{2} = \left( \pi^{2} \left( \frac{e}{g} \right)^{\frac{2}{2}}\right)
$$

Simplificando, obtenemos

$$
 P^{2} = \pi^{2} \left( \frac{e}{g} \right) \rightarrow P^{2} = \pi^{2} \cdot \frac{e}{g}
$$

Multiplicamos ambos lados de la igualdad por $g$

$$
 g\cdot P^{2} = g \cdot \pi^{2} \cdot \frac{e}{g}
$$

Obtenemos

$$
 g\cdot P^{2} = \pi^{2} \cdot e
$$

Multiplicamos ambos lados de la igualdad por $\frac{1}{P^2}$

$$
 \frac{1}{P^2} \cdot g\cdot P^{2} = \frac{1}{P^2} \cdot \pi^{2} \cdot e
$$

Finalmente, tenemos

$$
 g = \frac{\pi^{2} \cdot e}{P^2}
$$

2. Despeje $x$ de $y=\frac{1}{\sqrt{1-x}}$

Puedo escribirlo de la siguiente forma

$$
 y = \frac{1}{\left({1-x} \right)^{\frac{1}{2}}}
$$

Elevamos ambos lados de la igualdad al cuadrado y tenemos

$$
 y^{2} = \left( \frac{1}{\left({1-x} \right)^{\frac{1}{2}}} \right)^{2}
$$

Aplicando propiedades de los exponentes, obtenemos

$$
 y^{2} = \frac{1}{1-x}
$$

Multiplico ambos lados de la igualdad por $1-x$ 

$$
 (1-x) \cdot \left( y^{2} \right) = (1-x) \cdot \frac{1}{1-x} \rightarrow (1-x) \cdot \left( y^{2} \right) = 1,
$$

y después por $\frac{1}{y^2}$

$$
 \frac{1}{y^2} \cdot (1-x) \cdot \left( y^{2} \right) = \frac{1}{y^2} \cdot 1,
$$

Obtenemos entonces

$$
 1-x = \frac{1}{y^2}
$$

Restamos $1$ en ambos lados de la igualdad

$$
 1 -1 -x = \frac{1}{y^2} - 1
$$

Obtenemos

$$
 -x = \frac{1}{y^2} - 1
$$

Multiplicamos por $-1$ en ambos lados de la igualdad y ordenamos los términos

$$
 x = 1 - \frac{1}{y^2}
$$
