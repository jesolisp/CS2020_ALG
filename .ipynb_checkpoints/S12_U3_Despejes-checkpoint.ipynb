## Estructura de una ecuación

### Ecuaciones de primer grado

Las podemos identificar porque se encuentran en la forma canónica

$$
 ax + b = 0,
$$
donde la solución está dada por $-\frac{b}{a}$. Esta solución la podemos encontrar si aplicamos la **propiedad uniforme** o el **principio de transposición**.

#### Propiedad uniforme

$$
 \begin{aligned}
  ax + b + (-b) &= 0 + (-b) \\
  ax &= -b \\
  \frac{1}{a} ax &= \frac{1}{a} (-b) \\
  x &= -\frac{b}{a}
 \end{aligned}
$$

#### Principio de transposición

$$
 \begin{aligned}
  ax + b &= 0 \\
  ax &= -b \\
  x &= -\frac{b}{a}
 \end{aligned}
$$

**Ejemplo.** Resuelva la siguiente ecuación

$$
 x(x + 6) = (x + 4)(x - 3)
$$

1. Multiplicamos los binomios para obtener lo siguiente

$$
 x^{2} + 6x = x^{2} + x - 12
$$

2. Aplicando el principio de transposición, tenemos

$$
 x^{2} + 6x - x^{2} - x  = - 12
$$

3. Reduciendo términos semejantes

$$
 5x = -12 \rightarrow x = -\frac{12}{5}
$$

#### Ejemplo de aplicación

Cosme Fulanito entró a la Universidad Politécnica del estado de Morelos a estudiar la carrera de Ingeniería en Electrónica y Telecomunicaciones. Debido a la contigencia por el SARS-CoV-2 se vio en la necesidad de comprar un equipo de cómputo para tomar sus clases en línea. Desafortunadamente, la tienda donde adquirió su equipo le cobró $6\%$ de comisión y al final pagó $\$11,660.00$. ¿Cuánto le costó la computadora a Cosme Fulanito?

**Planteamiento**
Sea $x$ el precio de la computadora, entonces

$$
 x + 0.06x = 11660.
$$

**Resolución**

$$
 \begin{aligned}
  1x + 0.06x &= 11660 \\
  1.06x &= 11660 \\
  x &= \frac{11660}{1.06} \\
  x &= 11000.
 \end{aligned}
$$

**Comprobación**

$$
 \begin{matrix}
  Costo & 11000 \\
  Impuesto\ 6\% & 660 \\
  Total & 11660
 \end{matrix}
$$

### Ecuaciones de segundo grado
Este tipo de ecuaciones las podemos identificar recurriendo a su forma canónica

$$
 ax^{2} + bx + c = 0, \quad a\neq 0.
$$

En este tipo de ecuaciones tenemos lo siguiente:

1. Son completas si $b, c \neq 0$.
2. Sin incompletas si $b=0$ o $c=0$.

#### Métodos de resolución de ecuaciones de segundo grado

Una ecuación de segundo grado tiene dos soluciones o raíces y se pueden encontrar por medio de los siguientes métodos:

1. Factorización
2. Completar cuadrados
3. Método de la cuadrática

#### Solución por factorización

Sea $ax^{2} + bx + c = 0$, entonces $a(x - r_{1})(x - r_{2}) = 0$.

Aplicando la propiedad absorbente del cero

$$
 a\cdot 0 = 0.
$$

obtenemos

$$
 x - r_{1} = 0 \rightarrow x = r_{1},
$$

y

$$
 x - r_{2} = 0 \rightarrow x = r_{2},
$$

**Ejemplo 1.**
Considere la siguiente ecuación

$$
 x^{2} + 5x + 3 = 0
$$

Como $b,c \neq 0$, entonces la ecuación es **completa.** Esta ecuación, podemos factorizarla como sigue

$$
 (x + 1)(x + 2) = 0,
$$

de aquí podemos obtener

$$
 x + 1 = 0, \quad x + 2 = 0,
$$

y por lo tanto

$$
 x = -1, \quad x = -2;
$$

**Ejemplo 2.** Considere la siguiente ecuación

$$
 x^{2} - 4x = 0
$$

Como $c = 0$, entonces la ecuación es **incompleta.** Esta ecuación, podemos factorizarla como sigue

$$
 x(x - 4) = 0,
$$

de aquí podemos obtener

$$
 x = 0, \quad x - 4 = 0,
$$

y por lo tanto

$$
 x = 0, \quad x = 4;
$$

#### Solución por completación de cuadrados

Considere la siguiente ecuación

$$
 x^{2} + 3x - 10 = 0
$$

Utilizando el principio de transposición tenemos

$$
 x^{2} + 3x = 10
$$

Completar cuadrados consiste en agregar términos de modo que podamos tener un binomio cuadrado. Recordemos

$$
 (a + b)^{2} = a^{2} + 2ab + b^{2},
$$

entonces agregamos el término $\frac{9}{4}$ en ambos lados de la ecuación para que ésta no se vea afectada

$$
 x^{2} + 3x + \frac{9}{4} = 10 + \frac{9}{4}.
$$

Reducimos y tenemos

$$
 \left(x + \frac{3}{2}\right)^{2} = 10\cdot \frac{4}{4} + \frac{9}{4}.
$$

Recordando cómo sumar fracciones, tenemos

$$
 \left(x + \frac{3}{2}\right)^{2} = 10 + \frac{49}{4}.
$$

Para eliminar el expontente cuadrado, aplicamos la raíz cuadrada en ambos lados de la ecuación

$$
 \sqrt{\left(x + \frac{3}{2}\right)^{2}} = \sqrt{\frac{49}{4}}.
$$

Para obtener

$$
 \left(x + \frac{3}{2}\right) = \pm \frac{7}{2}.
$$

Reducimos términos semejantes

$$
 x = \pm \frac{7}{2} - \frac{3}{2}.
$$

De modo que

$$
 x = \frac{7}{2} - \frac{3}{2} = \frac{4}{2} = 2.
$$

y

$$
  x = - \frac{7}{2} - \frac{3}{2} = -\frac{10}{2} = -5.
$$

#### Solución por el método de la cuadrática

Considere la siguiente ecuación

$$
 3x^{2} - 10x - 8 = 0,
$$

1. Obtengamos los coeficientes $a,b$ y $c$ a partir de la siguiente representación

$$
 ax^{2} + bx + c = 0.
$$

Entonces,

$$
 a = 3, \quad b = -10, \quad c = -8.
$$

2. Recordando 

$$
 x = \frac{-b \pm \sqrt{b^{2} - 4ac}}{2a}
$$

3. Sustituyendo tenemos

$$
 \begin{aligned}
  x & = \frac{-(-10) \pm \sqrt{(-10)^{2} - 4(3)(-8)}}{2(3)}, \\
  & = \frac{10 \pm \sqrt{100 - 4(3)(-8)}}{6}, \\
  & = \frac{10 \pm \sqrt{100 + (12)(8)}}{6}, \\
  & = \frac{10 \pm \sqrt{100 + 96}}{6}, \\
  & = \frac{10 \pm \sqrt{196}}{6}, \\
  & = \frac{10 \pm 14}{6}, \\
 \end{aligned}
$$

donde tenemos las siguientes soluciones

$$
 x_{1} = \frac{10 + 14}{6} = \frac{24}{6} = 4,
$$

y

$$
 x_{1} = \frac{10 - 14}{6} = -\frac{4}{6} = -\frac{2}{3}.
$$

#### Nota:
En el método de la cuadrática, al radicando se le conoce como **discriminante** y éste se representa por

$$
 \triangle = b^{2} - 4ac.
$$

**Consideraciones**

1. Si $\triangle > 0$, entonces hay dos soluciones reales diferentes.
2. Si $\triangle = 0$, entonces hay una solución real de multiplicidad dos.
3. Si $\triangle < 0$, entonces no hay soluciones reales.

### Ejercicios de clase

Dadas las siguientes ecuaciones, despeje la variable indicada

1. Despeje $r$ para $c = 2\pi r$.
2. Despeje $a$ para $p = 2a + 2b$.
3. Despeje $n$ para $a_{n} = a_{1} + (n - 1)d$.

Resuelva la siguiente ecuación

$$
 x^{2}(3x + 5) = 4x + 4
$$

Dada la siguiente ecuación

$$
 4x^{2} - 4xy + 1 - y^{2} = 0
$$

aplicando el método de la cuadrática para obtener

1. $x$ en términos de $y$
2. $y$ en términos de $x$

Encuentre el valor de $k$ tal que el segundo polinomio sea un factor del primero

1. $x^{3} + x^{2} - 10x + k,\quad x - 4$.
2. $x^{4} + kx + 10,\quad x + 2$.
