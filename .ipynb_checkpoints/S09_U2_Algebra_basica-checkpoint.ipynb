## División de polinomios

Una de las formas para poder dividir polinomios es aplicando el algoritmo de Euclides.

Dados el dividendo $D(x)$ y el divisor $d(x)$ se puede obtener el cociente $q(x)$ y el residuo $r(x)$ tales que

$$
 grado\left\{r(x)\right\} < grado\left\{d(x)\right\}
$$

**Nota.** 
1. Si $r(x) = 0$, entonces la división es exacta.
2. Si $r(x) \neq 0$, entonces la división es inexacta.

Como tal no hay un consenso para factorizar polinomios pero bien nos podemos auxiliar de los productos notables vistos en la Sesión 7.

**Ejemplo**

Dados los siguientes binomios, factorice

* $2m^{3} + m$

Como podemos observar, el común en la expresión es la variable $m$. Por lo tanto, podemos factorizar esta expresión si consideramos a la variable $m$ como factor común. Esto es

$$
 m(2m^{2} + 1)
$$

* $4m + 2$

En este binomio, analizamos que el factor común es el número $2$. Por lo tanto, podemos factorizar de la siguiente manera

$$
 2(2m + 1)
$$

* $x^{4} + x^{2}y^{2} + y^{4}$

1. Agregamos el término $2x^{2}y^{2}$. Sin embargo, para que no se afecte el polinomio restamos $x^{2}y^{2}$. Es decir

$$
 x^{4} + 2x^{2}y^{2} - x^{2}y^{2} + y^{4}
$$

2. Ordenamos los términos

$$
 x^{4} + 2x^{2}y^{2} + y^{4} - x^{2}y^{2}
$$

3. Recordando productos notables

$$
 a^{2} + 2ab + b^{2} = (a + b)^{2}
$$

4. Recurrimos a una estrategia llamada **cambio de variable**. Consiste en asignarle a una variable un término con el fin de simplificar la operación.

$$
 a := x^{2},\quad b := y^{2},
$$
por consiguiente tenemos

$$
 a^{2} = x^{4}, \quad b^{2} = y^{4}.
$$

5. Entonces, el polinomio puede ser expresado como sigue

$$
 \left(x^{2} + y^{2} \right)^{2} - x^{2}y^{2}
$$

6. Nuevamente haciendo un cambio de variable, definimos como sigue

$$
 a := \left(x^{2} + y^{2} \right), \quad b := xy
$$
de modo que 

$$
 a^{2} = \left(x^{2} + y^{2} \right)^{2}, \quad b^{2} = x^{2}y^{2}.
$$

7. Recordando nuevamente los productos notables, tenemos

$$
 a^{2} - b^{2} = (a + b)(a - b).
$$

8. Sustituyendo

$$
 \left( x^{2} + y^{2} + xy \right)\left(x^{2} + y^{2} - xy \right)
$$

### ¿Para qué me sirve la división de polinomios?
**R.** Puede ser utilizada para simplificar. Por ejemplo, considere la siguiente división polinómica

$$
 \frac{x^{2} + 6x + 9}{x + 3}
$$

Podemos utilizar la división de la siguiente manera

$$
 x + 3 \sqrt{x^{2} + 6x + 9}
$$

Si utilizamos el algoritmo de división de polinomios, tenemos

![division_3](pictures/division_3.png)

Como el residuo es cero, decimos entonces que la división es exacta. Por lo tanto

$$
 \frac{x^{2} + 6x + 9}{x + 3} = \frac{(x + 3)(x + 3)}{(x + 3)} = x + 3.
$$

## División sintética o Regla de Ruffini

Considerado como el método para abreviar la división de un polinomio entre otro de primer grado o bien, un monomio.

**Ejemplo.** Considere dividir $D(x) = 2x^{4} - 17x^{2} + 6x -5$ entre $d(x) = x - 3$, entonces 

1. Tomamos $d(x) = 0$ y despejamos el valor de x, esto es $x - 3 = 0 \rightarrow x = 3$.
2. Llevamos el polinomio $D(x)$ a su forma canónica

$$
 2x^{4} + 0x^{3} - 17x^{2} + 6x -5
$$

3. Obtenemos los coeficientes

$$
 \begin{matrix}
  2 & 0 & -17 & 6 & -5
 \end{matrix}
$$

4. Construimos una "tabla" similar a la que sigue

![division_sintetica_1](pictures/division_sintetica_1.png)

5. Bajamos el número $2$

![division_sintetica_1](pictures/division_sintetica_2.png)

6. Después, el número $3$ multiplica al $2$ y el resultado será sumado con el $0$

![division_sintetica_1](pictures/division_sintetica_3.png)

![division_sintetica_1](pictures/division_sintetica_4.png)

7. Realizamos el mismo proceso, el número $3$ ahora multiplica al $6$ y el resultado será sumado con $-17$

![division_sintetica_1](pictures/division_sintetica_5.png)

![division_sintetica_1](pictures/division_sintetica_6.png)

8. Repetimos este procedimiento hasta obtener el residuo

![division_sintetica_1](pictures/division_sintetica_7.png)

![division_sintetica_1](pictures/division_sintetica_8.png)

![division_sintetica_1](pictures/division_sintetica_9.png)

![division_sintetica_1](pictures/division_sintetica_10.png)

Observamos que el residuo es $22$ y además, el polinomio resultante de la división se compone por los coeficientes obtenidos al final. Esto es

$$
 q(x) = 2x^{3} + 6x^{2} + x + 9,\ r = 22.
$$

#### ¿Qué pasa si tenemos la siguiente división?

Considere dividir $D(x) = 4x^{3} + 4x^{2} - 2$ entre $d(x) = 2x + 3$

1. Igualamos $d(x)=0$ y despejamos $x$

$$
 2x + 3 = 0 \rightarrow 2x = -3 \rightarrow x = -\frac{3}{2}
$$

2. Aplicando el mismo procedimiento tenemos lo siguiente

![division_4](pictures/division_4.png)

3. Tomando los coeficientes y el residuo obtenemos entonces

$$
 q(x) = 4x^{2} -2x + 3,\ r = -\frac{13}{2}
$$
