### Expresión algebráica

Compuesta por números, letras y signos de operaciones algebráicas. Por ejemplo:

$$
 Area = \pi r^{2}, \quad \frac{x^{2}}{1+x^{3}}, \quad x = \sqrt{x+y+z}
$$
donde los números son llamados **constantes** y las letras variables o **incógnitas**. La siguiente no es una expresión algebráica

$$
 \log{\left( \sin{\left( \frac{x}{6} \right)} \right)}
$$
ya que no contiene operaciones alegebráicas tales como $+,-,\cdot,/,\sqrt{\ }$.

Dentro de las expresiones racionales o irracionales tenemos el **monomio**.

### Monomio

Son de la forma $ax^{n}$ donde $a\in \mathbb{R}$ es llamado el **coeficiente** y $x^{n}$ la **parte literal** con un exponente $n\in \mathbb{N}$ que indica además el grado del monio. Por ejemplo:

$3x^{5}$ es un monomio de grado 5.

$\sqrt{2} = 2^{\frac{1}{2}}$ es un monomio de grado cero. ¿Por qué? porque toda constante $k\neq 0$ tiene grado cero.

### Monomios semejantes

Son aquellos que tienen la misma parte literal. Por ejemplo la siguiente expresión:

$$
 2x^{4} - \frac{1}{3}x^{4}.
$$

Como hay términos semejantes, el monomio se puede reducir a un solo monomio

$$
 \frac{5}{3}x^{4}.
$$

### Polinomios

Es la suma de monomios donde cada monomio es un término del polinomio, por ejemplo:

$$
 \begin{aligned}
  4x^{2} - 3x, \quad (binomio) \\
 2x^{3} + x^{2} - x,\quad (trinomio)
 \end{aligned}
$$

La representación canónica de un polinomio en $x$ sobre $\mathbb{R}$ está dado como sigue:

$$
 p(x) = a_{n}x^{n} + a_{n-1}x^{n-1} + \dots + a_{2}x^{2} + a_{1}x + a_{0},
$$
donde $a_{n},a_{n-1},\dots,a_{0} \in \mathbb{R}, a_{n} \neq 0$.

El $grado\{p(x)\} = n$, donde: $a_{n}x^{n}$ es el término principal y $a_{n}$ es el coeficiente principal.

**Ejemplo.** Considere el siguiente polinomio

$$
 p(x) = -4x^{3} - 2x^{2} + 3.
$$

Determine el grado de $p(x)$. Entonces, $grado\{p(x)\}=3$ dado que el máximo exponente es 3. Además, el coeficiente principal es $a_{n}=-4$ y $p(x)$ no es mónico. Tomando en cuenta los coeficientes, 

$$
 a_{2} = -2, a_{1} = 0, a_{0} = 3,
$$
entonces $p(x)$ se encuentra en forma canónica.

#### ¿Cómo determinamos si un polinomio está en forma canónica?

1. Debe estar ordenado en forma decreciente con respecto a los exponentes de su término.
2. Los términos semejantes son reducidos y los términos con coeficiente cero son omitidos.

## Sumas y restas de polinomios

Los términos semejantes son reducidos. En la resta, los signos del sustraendo se cambian y se suman al sustraendo.

**Ejemplo de suma** 

Sea el primer polinomio $p(x) = 2x^{3} - \frac{1}{4}x^{2} + 2$ y el segundo $g(x) = -6x^{3} + 3x + 3$, entonces $s(x) = p(x) + g(x)$ está dado como sigue

$$
 \begin{aligned}
  p(x) &= 2x^{3} - \frac{1}{4}x^{2} + 2 \\
  + \ \ \ \ \ \ \ \ & \\
  g(x) &= -6x^{3} + 3x^{2} + 3 \\
  --- & ---------- \\
 \end{aligned}
$$

Reducimos términos semejantes

$$
 \begin{matrix}
  & p(x) & = & 2x^{3} & - \frac{1}{4}x^{2} & + 2 \\
  + & & & & \\
  & g(x) & = & -6x^{3} & + 3x^{2} & + 3 \\
  & ---&---&---&---&--- \\
  & s(x) & = & -4x^{3} & + \frac{11}{4}x^{2} & + 5
 \end{matrix}
$$


Descomponemos el polinomio en monomios y sumamos cada uno de ellos

$$
 \begin{aligned}
  2x^{3} - 6x^{3} &= -4x^{3},\\
 -\frac{1}{4}x^{2} + 3x^{2} = -\frac{1}{4}x^{2} + 3\cdot\frac{4}{4}x^{2} &= -\frac{1}{4}x^{2} + \frac{12}{4}x^{2} = \frac{11}{4}x^{2},\\
 2 + 3 &= 5.
 \end{aligned}
$$

**Ejemplo de resta** 

Sea el primer polinomio $p(x) = 4x^{3} - \frac{1}{2}x^{2} + 4$ y el segundo $g(x) = -3x^{3} + 5x + 1$, entonces $r(x) = p(x) - g(x)$ está dado como sigue

$$
 \begin{aligned}
  p(x) &= 4x^{3} - \frac{1}{2}x^{2} + 4 \\
  - \ \ \ \ \ \ \ \ & \\
  g(x) &= -3x^{3} + 5x^{2} + 1 \\
  --- & ---------- \\
 \end{aligned}
$$

Multiplicamos $g(x)$ por $-1$ para obtener

$$
 \begin{aligned}
  p(x) &= 4x^{3} - \frac{1}{2}x^{2} + 4 \\
  -g(x) &= 3x^{3} - 5x^{2} - 1 \\
  --- & ---------- \\
 \end{aligned}
$$

Reducimos términos semejantes

$$
 \begin{matrix}
  p(x) & = & 4x^{3} & - \frac{1}{2}x^{2} & + 4 \\
  -g(x) & = & 3x^{3} & - 5x^{2} & - 1 \\
  ---&---&---&---&--- \\
  s(x) & = & 7x^{3} & - \frac{11}{2}x^{2} & + 3
 \end{matrix}
$$


Descomponemos el polinomio en monomios y sumamos cada uno de ellos

$$
 4x^{3} + 3x^{3} = 7x^{3},\\
 -\frac{1}{2}x^{2} - 5x^{2} = -\frac{1}{2}x^{2} - 5\cdot\frac{2}{2}x^{2} = -\frac{1}{2}x^{2} - \frac{10}{2}x^{2} = -\frac{11}{2}x^{2},\\
 4 - 1 = 3.
$$

#### ¿Cómo determinamos el grado del polinomio resultante?

$$
 grado\left\{p(x) \pm g(x) \right\} \leq \max{\left\{p(x),g(x) \right\}}
$$

### Ejercicios

1. Sea $p(x) = 3x^{3} + 2$ y $g(x) = -2x^{2} + x$ obtenga $s(x) = p(x) + g(x)$

$$
 \begin{aligned}
  p(x) &= 3x^{3} + 2 \\
  + \ \ \ \ \ \ \ \ & \\
  g(x) &= -2x^{2} + x \\
  --- & ---------- \\
 \end{aligned}
$$

Reducimos términos semejantes

$$
 \begin{matrix}
  & p(x) & = & 3x^{3} &  & & + 2 \\
  + & & & & \\
  & g(x) & = &  & - 2x^{2} & + x & \\
  & ---&---&---&---&---&---& \\
  & s(x) & = & 3x^{3} & -2x^{2} & + x & + 2
 \end{matrix}
$$

2. Sea $p(x) = \frac{1}{3}x^{5} + \frac{2}{8}x^{2} - 5$ y $g(x) = -\frac{1}{4}x^{5} + \frac{1}{6}x^{2} + 4x + 6$ obtenga $s(x) = p(x) + g(x)$

$$
 \begin{aligned}
  p(x) &= \frac{1}{3}x^{5} + \frac{2}{8}x^{2} - 5 \\
  + \ \ \ \ \ \ \ \ & \\
  g(x) &= -\frac{1}{4}x^{5} + \frac{1}{6}x^{2} + 4x + 6 \\
  --- & ------------- \\
 \end{aligned}
$$

Reducimos términos semejantes

$$
 \begin{matrix}
  & p(x) & = & \frac{1}{3}x^{5} & & & + \frac{2}{8}x^{2} & & - 5 \\
  + & & & & & & \\
  & g(x) & = & -\frac{1}{4}x^{5} & & & + \frac{1}{6}x^{2} & + 4x & + 6 \\
  &---&---&---&---&---&---&---&--- \\
  & s(x) & = & \frac{1}{12}x^{5} &  &  & + \frac{5}{12}x^{2} & +4x & +1 & 
 \end{matrix}
$$

Descomponemos el polinomio en monomios y sumamos cada uno de ellos

$$
 \frac{1}{3}x^{5} - \frac{1}{4}x^{5}
$$

Para poder sumar las fracciones anteriores, tenemos que recordar el **MCM**. Para ello, tomamos los denominadores y obtenemos sus múltiplos

$$
 \begin{aligned}
  3, 6, 9, \textbf{12}, 16, 18 \\
  4, 8, \textbf{12}, 16, 20
 \end{aligned}
$$

Entonces, el MCM es 12.

$$
 \frac{1}{3}x^{5} - \frac{1}{4}x^{5} = \frac{1}{3}\cdot \frac{4}{4}x^{5} - \frac{1}{4}\cdot \frac{3}{3}x^{5} = \frac{4}{12}x^{5} - \frac{3}{12}x^{5} = \frac{1}{12}x^{5} 
$$

Realizamos el mismo proceso para sumar las siguientes fracciones

$$
 \begin{aligned}
  \frac{2}{8}x^{2} + \frac{1}{6}x^{2} \\
  8, 16, \textbf{24}, 32, 40, 48, 56, 64 \\
  6, 12, 18, \textbf{24}, 30, 36, 42, 48, 50
 \end{aligned}
$$

Aquí el MCM es $\textbf{24}$. Entonces

$$
 \frac{2}{8}x^{2} + \frac{1}{6}x^{2} = \frac{2}{8}\cdot\frac{3}{3}x^{2} + \frac{1}{6}\cdot\frac{4}{4}x^{2} = \frac{6}{24}x^{2} + \frac{4}{24}x^{2} = \frac{10}{24}x^{2} = \frac{5}{12}x^{2}.
$$
